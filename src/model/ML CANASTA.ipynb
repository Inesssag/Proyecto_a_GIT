```python
import pandas as pd
from itertools import combinations
from collections import Counter

# Suponiendo que 'df' es tu DataFrame
# Agrupar por 'Movimiento PDV' y obtener los artículos en cada ticket
agrupado = df.groupby('Movimiento PDV')['Platillo / Artículo'].apply(list)

# Crear una lista de todas las combinaciones de tres artículos en cada ticket
combinaciones = []
for items in agrupado:
    if len(items) > 2:
        combinaciones.extend(combinations(sorted(items), 3))

# Contar la frecuencia de cada combinación
combinaciones_contador = Counter(combinaciones)

# Crear una lista de todos los artículos únicos
articulos_unicos = list(set(df['Platillo / Artículo']))

# Crear una matriz de co-ocurrencia (opcional, para otras visualizaciones)
co_ocurrencia = pd.DataFrame(0, index=articulos_unicos, columns=articulos_unicos)
for (articulo1, articulo2, articulo3), count in combinaciones_contador.items():
    co_ocurrencia.at[articulo1, articulo2] += count
    co_ocurrencia.at[articulo1, articulo3] += count
    co_ocurrencia.at[articulo2, articulo1] += count
    co_ocurrencia.at[articulo2, articulo3] += count
    co_ocurrencia.at[articulo3, articulo1] += count
    co_ocurrencia.at[articulo3, articulo2] += count

# Convertir la matriz a una lista de tríos únicos y ordenar por frecuencia
trios_frecuentes = [(articulo1, articulo2, articulo3) for (articulo1, articulo2, articulo3), count in combinaciones_contador.items()]
trios_frecuentes_ordenados = sorted(trios_frecuentes, key=lambda x: combinaciones_contador[x], reverse=True)

# Mostrar los tríos más frecuentes
print("Tríos de artículos más frecuentemente pedidos juntos:")
for articulo1, articulo2, articulo3 in trios_frecuentes_ordenados:
    print(f"{articulo1}, {articulo2} y {articulo3}")

```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

data1=  trios_frecuentes_ordenados

df_canasta3 = pd.DataFrame(data1, columns=['Item1', 'Item2', "Item3"])
print(df_canasta3)
```

```python
from mlxtend.preprocessing import TransactionEncoder

# Convertir las transacciones en el formato adecuado
te = TransactionEncoder()
te_ary = te.fit(data1).transform(data1)
df_canasta3 = pd.DataFrame(te_ary, columns=te.columns_)
print(df_canasta3)
```

```python
# Aplicar el algoritmo Apriori para encontrar itemsets frecuentes
frequent_itemsets = apriori(df_canasta, min_support=0.0003, use_colnames=True)
print(frequent_itemsets)
```

```python
# Generar reglas de asociación
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print(rules)
```

```python

def predict_third_product(product1, product2):
    # Filtrar reglas que contengan los dos productos dados en los antecedentes
    filtered_rules = rules[(rules['antecedents'].apply(lambda x: product1 in x and product2 in x))]
    if not filtered_rules.empty:
        # Ordenar por 'lift' para encontrar la regla más fuerte
        best_rule = filtered_rules.sort_values(by='lift', ascending=False).iloc[0]
        # Extraer el producto consecuente
        consequent = list(best_rule['consequents'])[0]
        return consequent
    else:
        return None
```

```python
product1 = 'CONCHA NATIVA'
product2 = 'CAFE REFILL'
predicted_product = predict_third_product(product1, product2)
print(f"Si alguien compra {product1} y {product2}, es probable que también compre: {predicted_product}")
```

```python
import pandas as pd
from itertools import combinations
from collections import Counter
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Suponiendo que 'df' es tu DataFrame
# Agrupar por 'Movimiento PDV' y obtener los artículos en cada ticket
agrupado = df.groupby('Movimiento PDV')['Platillo / Artículo'].apply(list)

# Crear una lista de todas las combinaciones de tres artículos en cada ticket
combinaciones = []
for items in agrupado:
    if len(items) > 2:
        combinaciones.extend(combinations(sorted(items), 3))

# Contar la frecuencia de cada combinación
combinaciones_contador = Counter(combinaciones)

# Crear una lista de tríos únicos sin la frecuencia
trios_frecuentes = [trio for trio, count in combinaciones_contador.items()]

# Convertir las transacciones en el formato adecuado
te = TransactionEncoder()
te_ary = te.fit(trios_frecuentes).transform(trios_frecuentes)
df_canasta3 = pd.DataFrame(te_ary, columns=te.columns_)
print(df_canasta3)

# Aplicar el algoritmo Apriori para encontrar itemsets frecuentes
frequent_itemsets = apriori(df_canasta3, min_support=0.0003, use_colnames=True)
print(frequent_itemsets)

# Generar reglas de asociación
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print(rules)

def predict_third_product(product1, product2):
    # Filtrar reglas que contengan los dos productos dados en los antecedentes
    filtered_rules = rules[rules['antecedents'].apply(lambda x: product1 in x and product2 in x)]
    if not filtered_rules.empty:
        # Ordenar por 'lift' para encontrar la regla más fuerte
        best_rule = filtered_rules.sort_values(by='lift', ascending=False).iloc[0]
        # Extraer el producto consecuente
        consequent = list(best_rule['consequents'])[0]
        return consequent
    else:
        return None

product1 = 'AVOCADO TOAST'
product2 = 'AGUACATE'
predicted_product = predict_third_product(product1, product2)
print(f"Si alguien compra {product1} y {product2}, es probable que también compre: {predicted_product}")
```
