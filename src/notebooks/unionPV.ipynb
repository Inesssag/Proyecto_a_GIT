```python
archivos_excel=glob.glob(f"{carpeta}/*.xlsx")
```

```python
if not archivos_excel:
    raise FileNotFoundError("No se encontraron archivos Excel (.xlsx) en la carpeta especificada.")
```

```python
dataframes = []
```

```python
for archivo in archivos_excel:
    try:
        # Leer todas las hojas de cálculo del archivo Excel
        xls = pd.ExcelFile(archivo)
        for nombre_hoja in xls.sheet_names:
            # Leer cada hoja de cálculo y agregarla a la lista de DataFrames
            df = pd.read_excel(xls, sheet_name=nombre_hoja)
            dataframes.append(df)
    except Exception as e:
        print(f"Error al leer el archivo {archivo}: {e}")
```

```python
# Mostrar el número total de hojas de cálculo leídas
print(f"Se leyeron un total de {len(dataframes)} hojas de cálculo.")
```

Puedes acceder a cada DataFrame individualmente en la lista `dataframes`

```python
# DataFrames para almacenar las primeras y segundas hojas de cada archivo
primeras_hojas = []
segundas_hojas = []
```

```python
# Leer la primera y segunda hoja de cada archivo Excel
for archivo in archivos_excel:
    try:
        # Leer la primera hoja de cálculo
        primer_df = pd.read_excel(archivo, sheet_name=0)
        primeras_hojas.append(primer_df)

        # Leer la segunda hoja de cálculo si existe
        xls = pd.ExcelFile(archivo)
        if len(xls.sheet_names) > 1:
            segundo_df = pd.read_excel(xls, sheet_name=1)
            segundas_hojas.append(segundo_df)
    except Exception as e:
        print(f"Error al leer el archivo {archivo}: {e}")
```

```python
# Concatenar todos los DataFrames de las primeras hojas en uno solo
df_primeras_hojas = pd.concat(primeras_hojas, ignore_index=True)
```

```python
# Concatenar todos los DataFrames de las segundas hojas en uno solo
df_segundas_hojas = pd.concat(segundas_hojas, ignore_index=True)
```
